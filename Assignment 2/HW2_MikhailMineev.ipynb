```python
import numpy as np

#Calculated the gcd
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

#generate d
def multInv(e, phi):
    for i in range(1, phi):
        if (e * i) % phi == 1:
            return i
    return 

#returns True if num is prime False if not
def is_prime(num):       
    if num == 1:
        return False
    if num == 2:
        return True
    else:
        for i in range(2, int(num/2)+1):
            if(num%i) == 0:
                return False
            else:
                return True

#checks if entered nums are prime
#checks if entered nums are the same
def checkInt(p,q):
    if(is_prime(p) == False):
        raise ValueError("First number is not prime")
    if(is_prime(q) == False):
        raise ValueError("Second number is not prime")
    if(p == q):
        raise ValueError("Numbers cannot be equal")
    
#find the coprime of the entered num   
def coPrimes(ent):
    l = []
    for i in range(2, ent):
        if gcd(ent, i) == 1: 
            if multInv(i,ent) != None:
                l.append(i)
    return l
    
#generates the public and private keys
def generate_keypair(p, q):
    n = p * q 
    phi = (p-1) * (q-1)
    e = int(np.random.choice(coPrimes(phi)))
    d= multInv(e,phi)
   
    return ((e, n), (d, n))


#encrypt the message in file.txt
#output to encryptedfile.txt
def encryptmessage(file, keyType):
    arr = file.readlines()
    plaintext = ""
    
    for i in arr:
        plaintext = i.replace("\n", " ")
        key,n = keyType
        
        cypherText = [(ord(char) ** key) % n for char in plaintext]
        wf = " ".join(map(lambda x: str(x), cypherText))
        
        writetofile = open("encryptedfile.txt", "w")
        writetofile.write(wf)

#decrypt the message in encryptedfile.txt
#output to decryptedfile.txt
def decryptmessage(file, keyType):

    encrypted = file.read()
    encr = (int(i) for i in encrypted.split())
    key, n = keyType
    
    plainText = [chr((char ** key) % n) for char in encr]
    writetofile = open("decryptedfile.txt", "w")
    writetofile.write("".join(plainText))

    
    
    
#########################################    

p = int(input("Enter a prime number"))
q = int(input("Enter a different prime number"))
checkInt(p,q)
public, private = generate_keypair(p, q)


file1 = open("file.txt","r")
encryptmessage(file1, private)

file2 = open("encryptedfile.txt","r")
decryptmessage(file2, public)


print("This is the public key:" + str(public))
print("This is the private key:" + str(private))
file1 = open("file.txt","r")
print("This is our original message: " + file1.read())
file2 = open("encryptedfile.txt","r")
print("This is the encrypted message: " + file2.read())
file3 = open("decryptedfile.txt","r")
print("This is our decrypted message: " + file3.read())

```
