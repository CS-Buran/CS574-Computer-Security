```python
import numpy as np
import binascii

# Read in the binary from the text file
def readF():
    file = open("HW3_twotimepad.txt","r")
    l1 = file.read().split('\n')[0]
    file.close()
    c1 = l1[7:]
    
    file = open("HW3_twotimepad.txt","r")
    l2 = file.read().split('\n')[2]
    file.close()
    c2 = l2[7:]
    
    file = open("HW3_twotimepad.txt","r")
    l3 = file.read().split('\n')[4]
    file.close()
    c3 = l3[7:]
            
    file = open("HW3_twotimepad.txt","r")
    l4 = file.read().split('\n')[6]
    file.close()
    c4 = l4[7:]
    
    
    return c1,c2,c3,c4
    

#convert binary to hex
def toHex():
    c1,c2,c3,c4 = readF()
    
    hex1 = hex(int(c1, 2))
    hex2 = hex(int(c2, 2))
    hex3 = hex(int(c3, 2))
    hex4 = hex(int(c4, 2))
    
    return hex1,hex2,hex3,hex4

#xor message 1 and 2, and message 3 and 4.    
def xorS():
    a,b,c,d = toHex()
    
    x1 = hex(int(a,16) ^ int(b,16))
    x2 = hex(int(c,16) ^ int(d,16))
    
    
    return x1,x2
    
def cribD1(string):
    x1,x2 = xorS()
    
    x0 = string
    hexVal = x0.encode().hex()
    #comp = hex( int(hexVal,16) ^ int(x1,16))
    lengthVal = len(hexVal)
    for x in range(len(x1)-2):
        res = hex(int(x1[x+2:x+lengthVal+2],16) ^ int(hexVal,16))
        print(x1[x+2:x+lengthVal])
        print(res)
        print(x, "In x1:", res)
    return

cribD1("the")
```
